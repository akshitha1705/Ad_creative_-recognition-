```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

# Load the trained model
model = load_model(r'/kaggle/input/best-model/best_model (1).h5')

def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0  # Normalize the image to the range [0, 1]
    return img_array

def make_prediction(model, img_path):
    img_array = preprocess_image(img_path)
    predictions = model.predict(img_array)
    return predictions

def interpret_binary_predictions(predictions, classes, threshold=0.5):
    predicted_class_idx = (predictions > threshold).astype(int)[0][0]
    predicted_class_name = classes[predicted_class_idx]
    confidence = predictions[0][0] if predicted_class_idx == 1 else 1 - predictions[0][0]
    print(f"The predicted class is: {predicted_class_name} with confidence {confidence * 100:.2f}%")

# Define the classes
classes = np.array(['Non-Advertising Images', 'Advertising Images'])

# Example usage
img_path = r'/kaggle/input/images/img1.jpeg'
predictions = make_prediction(model, img_path)
interpret_binary_predictions(predictions, classes)
```
